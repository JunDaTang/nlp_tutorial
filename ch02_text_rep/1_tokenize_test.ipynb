```python
import jieba
```

```python
text = "小明毕业于北京大学计算机系"
```

## 精确模式（默认）

```python
# 返回一个词生成器对象
generator = jieba.cut(text)
print(generator)
```

```python
for word in generator:
    print(word)
```

```python
# 返回一个词的列表
word_list = jieba.lcut(text)
print(word_list)
```

## 全模式

```python
# 返回一个词生成器对象
generator = jieba.cut(text, cut_all=True)
print(generator)
for word in generator:
    print(word)
```

```python
# 返回一个词的列表
word_list = jieba.lcut(text, cut_all=True)
print(word_list)
```

## 搜索引擎模式

```python
# 返回一个词生成器对象
generator = jieba.cut_for_search(text)
print(generator)
for word in generator:
    print(word)
```

```python
# 返回一个词的列表
word_list = jieba.lcut_for_search(text)
print(word_list)
```

```python
text = "武汉市长江大桥"
word_list = jieba.lcut(text)
print(word_list)
```
