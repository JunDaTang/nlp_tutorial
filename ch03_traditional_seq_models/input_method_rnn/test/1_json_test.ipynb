```python
import pandas as pd
```

## 将DataFrame转换为JSON，保存成文件

```python
# 定义DataFrame
df = pd.DataFrame(
    dict(A=range(1, 4), B=range(4, 7), C=range(7, 10)),
    columns=list("ABC"),
    index=list("xyz"),
)
print(df)
```

```python
# 转成JSON文件
df.to_json("test.json", lines=True, orient="records")
```

## 读取JSON文件，转换为DataFrame

```python
df2 = pd.read_json("test.json", lines=True, orient="records")
print(df2)
```

```python
# 转成字典形式，方便代码处理
print(df2.to_dict(orient="records"))
```
