```python
import torch
import torch.nn as nn
```

```python
# 1. 定义输入数据：形状 (N, L, m)
input = torch.randn(2, 4, 3)
print(input.shape)
```

```python
# 2. 定义模型（LSTM层，两层双向）
lstm = nn.LSTM(input_size=3, hidden_size=5, bidirectional=True, batch_first=True, num_layers=2)
```

```python
# 3. 前向传播，得到输出
h0 = torch.randn(4, 2, 5)
c0 = torch.randn(4, 2, 5)
output, (hn, cn) = lstm(input, (h0, c0))
print(output.shape)     # (N, L, 2 × hidden_size)
print(hn.shape)         # (2×2, N, hidden_size)
print(cn.shape)
```
