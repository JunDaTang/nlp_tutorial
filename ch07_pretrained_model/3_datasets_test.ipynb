## 数据预处理

```python
print(dataset)
```

```python
# 1. 删除列
dataset = dataset.remove_columns(['cat'])
print(dataset)
```

```python
# 2. 过滤行
dataset = dataset.filter(lambda x: x['review'] is not None)
```

```python
print(dataset)
```

```python
# 3. 划分数据集
dataset_dict = dataset.train_test_split(test_size=0.2)
```

```python
print(dataset_dict)
```

```python
dataset_dict['train'][0]
```

```python
# 4. 编码数据：map方法与 tokenizer 配合
from transformers import AutoTokenizer
```

```python
# 4.1 创建分词器
tokenizer = AutoTokenizer.from_pretrained("google-bert/bert-base-chinese")
```

```python
# 4.2 定义数据集
train_dataset = dataset_dict['train']
print(train_dataset)
```

```python
# 4.3 编码转换，单样本、批量样本通用
def batch_encode(example):
    inputs = tokenizer(
        example['review'],
        padding="max_length",
        max_length=128,
        truncation=True,
    )
    # 添加标签字段labels
    inputs['labels'] = example['label']
    return inputs

train_tensor_dataset = train_dataset.map(batch_encode, batched=True, remove_columns=['label', 'review'])
```

```python
print(train_tensor_dataset)
```

```python
print(train_tensor_dataset[0])
```

## 集成DataLoader

```python
from torch.utils.data import DataLoader
```

```python
# 从文件加载训练数据集
train_dataset = load_from_disk("./data/processed/train")
```

```python
print(train_dataset)
```

```python
train_dataset.set_format(type='torch')
```

```python
# 创建DataLoader
dataloader = DataLoader(train_dataset, batch_size=32, shuffle=True)
```

```python
# 取一批
data_iter = iter(dataloader)
batch = next(data_iter)
print(batch)
```
