## 加载模型

```python
from transformers import AutoModel, AutoModelForSequenceClassification
```

```python
# 加载 BERT 模型
model = AutoModel.from_pretrained("google-bert/bert-base-chinese")
```

```python
print(model)
```

```python
print(type(model))
```

```python
# 从本地加载模型
model = AutoModel.from_pretrained("./pretrained/bert-base-chinese")
```

```python
print(type(model))
```

## 加载带任务头的模型

```python
from transformers import AutoModelForSequenceClassification
```

```python
# 创建模型
model = AutoModelForSequenceClassification.from_pretrained("google-bert/bert-base-chinese", num_labels=5)
```

```python
print(model)
```
