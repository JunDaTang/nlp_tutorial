```python
import torch
from torch.nn.utils.rnn import pad_sequence
```

```python
# 定义数据：3个长度不同的序列
seq1 = torch.randint(1000, (10, ))
seq2 = torch.randint(1000, (5, ))
seq3 = torch.randint(1000, (13, ))
print(seq1)
print(seq2)
print(seq3)
```

```python
# 将3条数据合并成一个batch
seq_list = [seq1, seq2, seq3]
print(seq_list)
```

```python
seq_batch = pad_sequence(seq_list, batch_first=True, padding_value=10)    
print(seq_batch)
print(seq_batch.shape)
```
