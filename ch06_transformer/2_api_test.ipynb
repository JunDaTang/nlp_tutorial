```python
import torch.nn as nn
```

## 模型定义

```python
# 定义 Transformer 模型
model = nn.Transformer(
    d_model=64,
    nhead=8,
    num_encoder_layers=3,
    num_decoder_layers=3,
    dim_feedforward=256,
    batch_first=True,
)
```

```python
print(model)
```

## 前向传播

```python
import torch
```

```python
# 定义源序列和目标序列
src = torch.randn(32, 10, 64)
tgt = torch.randn(32, 24, 64)
```

```python
# 前向传播
output = model(src, tgt)
print(output.shape)
```

## 编码

```python
memory = model.encoder(src)
print(memory.shape)
```

## 解码

```python
output = model.decoder(tgt, memory)
print(output.shape)
```
