```python
import math
import torch
```

```python
# torch.set_printoptions(sci_mode=False, precision=6)
torch.manual_seed(42)
```

```python
SEQ_LEN = 4
dk = 64

# 定义 Q、K、V
Q = torch.randn(size=(SEQ_LEN, dk), requires_grad=True)
K = torch.randn(size=(SEQ_LEN, dk), requires_grad=True)
V = torch.randn(SEQ_LEN, dk)
```

```python
# 按照公式计算注意力评分（点积评分）
scores = Q @ K.T / math.sqrt(dk)
print(scores)
```

```python
# 计算注意力权重
weights = torch.softmax(scores, dim=-1)
print(weights)
```

```python
# 模拟一个损失函数，反向传播
weights[0, 0].backward()
```

```python
# 查看梯度
print(Q.grad)
print(K.grad)
```

```python
# 计算上下文向量
context_vector = weights @ V
print(context_vector)
print(context_vector.shape)
```
